```python
#creation des trois symetries etudiees
from pymatgen.core.operations import SymmOp as So
from IPython.display import Image
from IPython.display import Math

Sym1 = So.from_xyz_string('-y,x-y,z')
Sym2 = So.from_xyz_string('-x+y,y,z')
Sym3 = So.from_xyz_string('y+1/3, x-1/3,-z-1/3')

#affichage de la premiere symetrie

print('\n')
print("symétrie 1 : symétrie de rotation d'ordre 3 (120°) selon l'axe c ")
print('matrice de rotation :')
print(Sym1.rotation_matrix)
print('vecteur de translation : ')
print(Sym1.translation_vector)
print("Appliqué à l'atome S2 en {0,0,0.26} on obtient les nouvelles coordonnées : ")
newc_1 = Sym1.operate((0.0,0.0,0.26))
print(newc_1)
print('\n')
print("Illustrations de la symétrie C3 appliquée à l'atome S2 :")
Image("Sym1(2).jpg")


```

```python
%%latex

Système d'équations décrivant la 1ere symétrie 

\begin{equation}
\begin{pmatrix} 0 & -1 & 0 \\1 & -1 & 0  \\0 & 0 & 1 \end{pmatrix}\begin{pmatrix}x \\ y \\ z\end{pmatrix}= \begin{pmatrix}-y \\ x-y \\ z \end{pmatrix}
\end{equation}
```

```python
print('\n')
print("symétrie 2 : plan miroir ")
print('matrice de rotation :')
print(Sym2.rotation_matrix)
print('vecteur de translation : ')
print(Sym2.translation_vector)
print("Appliqué à l'atome Na0 en {2/3,1/3,5/6} on obtient les nouvelles coordonnées : ")
newc_2 = Sym2.operate((2/3,1/3,5/6))
print(newc_2)
print('\n')
print("Illustration de la symétrie plan miroir appliquée à l'atome Na0 (le plan de symétrie est en vert):")
Image("Sym2(2).jpg")
```

```python
%%latex

Système d'équation décrivant la 2ème symétrie :

\begin{equation}
\begin{pmatrix} -1 & 1 & 0 \\ 0 & 1 & 0 \\ 0&0&1 \end{pmatrix}
\begin{pmatrix} x \\ y \\ z \end{pmatrix} =
\begin{pmatrix} -x + y \\ y \\ z \end{pmatrix}

\end{equation}
```

```python
print('\n')
print("symétrie 3 : rotation d'axe C d'ordre 2 ")
print('matrice de rotation :')
print(Sym3.rotation_matrix)
print('vecteur de translation : ')
print(Sym3.translation_vector)
print('\n')
print("Appliqué à l'atome Er1 en {2/3,1/3,1/3} on obtient les nouvelles coordonnées : ")
newc_3 = Sym3.operate((2/3,1/3,1/3))
print(newc_3)
print('\n')
print("Illustrations de la symétrie C2 appliquée à l'atome Er1 :")
Image("Sym3(3).jpg")
```
